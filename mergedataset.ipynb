```python
import pandas as pd
df1 = pd.read_csv('CyberUAVIntrusion1.csv')
df2 = pd.read_csv('CyberUAVIntrusion2.csv')
df1_cols = set(df1.columns)
df2_cols = set(df2.columns)
for col in df2_cols - df1_cols:
    df1[col] = pd.NA

for col in df1_cols - df2_cols:
    df2[col] = pd.NA
df2 = df2[df1.columns]
merged_df = pd.concat([df1, df2], ignore_index=True)
merged_df.to_csv('CyberUAVIntrusion.csv', index=False)
print("Shape after merging:", merged_df.shape)
merged_df.head()
```
